```r
# Arithmetic operators are +, -, /, *, %, ^. 
# All are same as python. So self explanatory.

# Vectors.
numbers = c(3, 5, 12, 0, 11) # numbers = [3, 4, 12, 0, 11]
print(numbers)
print(numbers[1]) # Not 0 based like Python!
print(length(numbers)) # len(numbers) in python.
print(numbers[5]) # Last element
print(numbers[6]); # Nice!
print(numbers[c(1, 2)]) # Not that great. numbers[:2] in python.
numbers[c(1, 2)] = 808 # Changing multiple elements in a list.
print(numbers)
print(numbers[-1]) # To omit first element.
print(numbers[2:3])
numbers[3] = 999
print(numbers)
print(sort(numbers, decreasing = F))
print(sort(numbers, decreasing = T))
"------------------"
print(1:19)
print(seq(from = 3, to = 93, by = 10))
print(seq(from = 3, to = 93, by = 10))
sample = seq(from = 3, to = 93, by = 10)
sprintf("13 in sample: %s", 13 %in% sample) # Example print statement that is more useful. Python: 13 in sample, done!
"------------------"
print(rep(x = 3))
print(rep(x = 3, times = 5))
print(rep(x = 3, times = 5, each = 2))
print(rep(x = c(1, 2, 3), times = 5, each = 2)) 
```
